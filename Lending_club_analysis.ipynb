```python
loan_df = loan_df.drop(['tax_liens','delinq_amnt','chargeoff_within_12_mths','acc_now_delinq','application_type','policy_code'
             ,'collections_12_mths_ex_med','pymnt_plan','initial_list_status','next_pymnt_d','pub_rec_bankruptcies'],axis=1)
```

A FEW MORE COLUMNS like zip_code, out_prnp ARE NOT REQUIRED FOR THE ANALYSIS SO REMOVING THEM

```python
loan_df.columns
```

```python
loan_df = loan_df.drop(['zip_code','delinq_2yrs','mths_since_last_delinq','mths_since_last_record','out_prncp'
             ,'out_prncp_inv','last_pymnt_d','last_credit_pull_d'],axis=1)
```

```python
loan_df
```

```python
plt.boxplot([loan_df.annual_inc]) #Check for the outliers
```

 HERE WE CAN SEE OUTLIERS PRESENT, SO REMOVE THEM

```python
Percentile = np.percentile(loan_df.annual_inc,[5,95])
loan_df = loan_df[(loan_df.annual_inc>=Percentile[0]) & (loan_df.annual_inc<=Percentile[1])]
```

```python
plt.boxplot([loan_df.annual_inc])
```

HERE WE CAN OBSERVE OUTLIERS ARE REMOVED

```python
loan_df.isnull().sum()
```

HERE WE CAN SEE emp_length COLUMN HAVE SOME NULL VALUES SO REPLACE THEM WITH MODE

```python
loan_df.emp_length.value_counts()
```

FROM THE ABOVE VALUES WE CAN SEE THAT 10+ YEARS HAS A VERY FAR VALUE SO WE CAN REPLACE THAT FOR REMAINING NULL VALUES

```python
loan_df.emp_length = loan_df.emp_length.fillna(loan_df.emp_length.mode()[0])
```

```python
loan_df.isnull().sum()
```

HERE WE CAN SEE revol_util COLUMN HAVE SOME NULL VALUES SO REPLACE THEM WITH MODE

```python
loan_df.revol_util.value_counts()
```

FROM THE ABOVE VALUES WE CAN SEE THAT 0% HAS A VERY FAR VALUE SO WE CAN REPLACE THAT FOR REMAINING NULL VALUES SO THAT THE ASSUMED VALUE DOESN'T AFFECT OUR ANALYSIS.

```python
loan_df.revol_util = loan_df.revol_util.fillna(loan_df.revol_util.mode()[0])
```

```python
loan_df.isnull().sum()
```

# NOW let us start the analysis

At first, look at the count of charged off vs fully paid borrowers

```python
#count of charged off vs fully paid borrowers
loan_df['loan_status'].value_counts().plot.pie()
```

```python
plt.hist(loan_df.loan_status)
```

```python
loan_df.loan_status.value_counts(normalize=True)
```

Above chart shows ~14% of the people charged off among all the borrowers which is ~5k of them.

Loan amount is the amount of loan applied by the borrower and funded amount is the amount we assume is the amount sanctioned.

```python
#Display funded amount vs count of Charged off members
loan_df['funded_amount_bins'] = pd.cut(loan_df['funded_amnt'], bins=4,labels=['0-10000','10000-20000','20000-30000','30000-40000'])
sns.countplot(x='funded_amount_bins',data=loan_df[loan_df.loan_status == 'Charged Off'])
```

Based on the above plot, borrowers with lesser funded amount have charged off more.

Term is the duration of loan repayment period of borrowers, It is 2 terms here, 3 years and 5years

```python
#analyse term vs charged off
sns.displot(x='term',data=loan_df[loan_df.loan_status == 'Charged Off'])
```

From the above plot, borrowers with 3 years loan repayment term charged off more than that of whom with 5 years term.

Interest rate is the percent of amount a borrower is imposed on the amount sanctioned.

```python
#Analyse interest rates with that of charged off persons
loan_df['int_rate_bins'] = pd.cut(loan_df['int_rate'], bins=4,labels=['0-10','10-20','20-30','30-40'])
sns.countplot(x='int_rate_bins',data=loan_df[loan_df.loan_status == 'Charged Off'])
```

From the above plot,when interest rate is more than 10% charged off count is much higher, so Lending Club should be looking for amendments on the interest rate and to keep it as lower as possible.

Home ownership is the status whether the borrower owns a housing property or are they on rent or on mortgage.

```python
#Analyse home ownership status with that of charged off people.
sns.displot(x='home_ownership',data=loan_df[loan_df.loan_status == 'Charged Off'])
```

From the above plot, borrowers who own a property charged off very much lesser that who rent or on mortgage.

Verifications status is assumed to be the status of the verification done by the Lending Club while issuing the funds for borrower.

```python
#Analyze verification status done on borrower with that of charged off status.
sns.displot(x='verification_status',data=loan_df[loan_df['loan_status'] == 'Charged Off'])
```

Based on the above plot, we assume Verified and Source verified as same and their combined count is greater than that of not verified status, so people whose background verification status is marked as verified, they charged off more which means the verification processes should be done properly by LC going forward.

dti is the debt to income ratio which means the amount of outstanding debt a borrower has to pay to income that is annual income.

```python
#analyse dti to charged off status
loan_df['dti_bins'] = pd.cut(loan_df['dti'], bins=6,labels=['0-5','5-10','10-15','15-20','20-25','25-30'])
sns.countplot(x='dti_bins',data=loan_df[loan_df['loan_status'] == 'Charged Off'])
```

Based on the above plot, when dti ratio is higher charged off rate is also higher, so when the dti ratio is lesser which means loan amount should be sanctioned in lower amounts based on their annual income.

Installment is the monthly amount being re-paid by borrowers.

```python
#analyse installment amount with charged off
loan_df['installment_bins'] = pd.cut(loan_df['installment'], bins=7,labels=['0-200','200-400','400-600','600-800','800-1000','1000-1200','1200-1400'])
sns.countplot(x='installment_bins',data=loan_df[loan_df.loan_status == 'Charged Off'])
```

When installment amount is lower borrowers charged off more, which might be due to their negligence to re-pay.
